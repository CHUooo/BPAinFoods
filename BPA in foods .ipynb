```python
bpa_g = bpa.groupby('Food_group')['mean_conta_UB'].max().sort_values(ascending=False).round(2)
```

```python
bpa_g
```

```python
bpa_20 = bpa_g.iloc[0:7]
```

```python
chart_1 = bpa_20.sort_values(ascending=True).plot(kind='barh', color='lightseagreen', alpha=1, figsize=(5, 7))
chart_1.set_xlim(11, 50)
chart_1.set_title('Foods in the Double Digit BPA Concentration')
chart_1.set_ylabel('')
chart_1.set_xlabel('Max Concentration Amount within group')
```
